```python
import pandas as pd

# Cargar los datos
df = pd.read_csv("../vehicles_us.csv")

# Mostrar las primeras 5 filas
df.head()
```

```python
# Información general del dataset
df.info()
# Resumen estadístico de las variables numéricas
df.describe()
# Contar valores únicos en la columna 'type' (tipo de vehículo)
df["type"].value_counts()
```

```python
import plotly.express as px

fig = px.histogram(df, x="price", title="Distribución de Precios de Vehículos")
fig.show()
```

```python
fig = px.scatter(df, x="model_year", y="price", color="type", title="Relación entre Año y Precio")
fig.show()
```

```python
fig = px.histogram(df, x="odometer", title="Distribución de Kilometraje")
fig.show()
```
